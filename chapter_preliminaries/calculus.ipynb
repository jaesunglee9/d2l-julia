# 2.4.1. Derivatives and Differentiation

```julia
using Printf, Plots
```

```julia
f(x) = 3x^2 - 4x
```

```julia
for h in [10. .^ h for h in -1:-1:-6]
    println(@sprintf("h=%.5f, numerical limit = %.5f", h, (f(1+h)-f(1))/h))
end
```

# 2.4.2. Visualization Utilities

```julia
x = 0:0.1:3
plot(x, [f.(x) 2x .- 3], xlabel="x", ylabel="f(x)", label=["f(x)" "Tangent line (x=1)"])
```
